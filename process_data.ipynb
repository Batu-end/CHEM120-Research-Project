```python
# Make a copy of both dataframes to be safe
fall_df_ready = fall_df.copy()
old_df_ready = old_df.copy()

# Combine them vertically (stack them with concatination)
# ignore_index=True resets the row numbers (0, 1, 2...) for the combined table
combined_df = pd.concat([old_df_ready, fall_df_ready], ignore_index=True)

# save into new Excel file
combined_df.to_excel('data/Combined_Data.xlsx', index=False)

# Pandas shape
print("Combined Data Shape (Rows, Columns):", combined_df.shape)
print("\nFirst 5 rows of Combined Data:")
display(combined_df.head())
```
