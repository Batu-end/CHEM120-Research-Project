# Data Processing
Adding IDs to the Old Dateset to match the format of the Fall Dataset.

```python
import pandas as pd

fall_df = pd.read_excel('Fall_Data.xlsx')
old_df = pd.read_excel('Old_Data.xlsx')

print("Fall Data columns:")
print(fall_df.columns.tolist())

print("\nOld Data columns:")
print(old_df.columns.tolist())
```

```python
# Load the old dataset
old_df = pd.read_excel('Old_Data.xlsx')

# Add 'Id' column starting from 77, 78, 79...
starting_id = 77
ids = list(range(starting_id, starting_id + len(old_df)))

# Insert the 'Id' column at the beginning (index 0) to match Fall_Data
old_df.insert(0, 'Id', ids)

# Save the updated dataframe back to Old_Data.xlsx
old_df.to_excel('Old_Data.xlsx', index=False)

print(f"Added 'Id' column to Old_Data.xlsx. Range: {ids[0]} to {ids[-1]}")
```
